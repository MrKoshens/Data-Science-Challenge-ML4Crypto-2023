```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from tqdm import tqdm
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/kunal/Downloads/TrainingData.csv')
```

```python
df.drop(['CID'], axis=1)
```

```python
# Expand the Bitstream column into multiple columns
expanded_bitstream = df['Bitstream'].apply(lambda x: pd.Series(list(x)))

# Concatenate the expanded Bitstream columns with the class column
df = pd.concat([expanded_bitstream, df['class']], axis=1)
```

```python
df
```

```python
X = df.drop(['class'], axis=1)

y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
estimators = [('svc', SVC(C=1, probability=True))]
```

```python
voting_classifier = VotingClassifier(estimators=estimators, voting='soft')
```

```python
# Train the voting classifier
voting_classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = voting_classifier.predict(X_test)
```

```python
# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
